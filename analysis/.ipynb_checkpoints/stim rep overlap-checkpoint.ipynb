```python
rule_set = ['fdgo', 'reactgo', 'delaygo', 'fdanti', 'reactanti', 'delayanti',
              'delaydm1', 'delaydm2','dmsgo', 'dmsnogo', 'dmcgo', 'dmcnogo']
task_num_set = np.concatenate((range(6),range(11,13),range(16,20)))

dir_set = ['lowD/grad_norm_l2001', 'lowD/combos', 'lowD/grad_norm_l2h000001']
n_components = 50

for dir_specific_all in dir_set:
    
    fldr = os.path.join('stim rep overlap/',dir_specific_all)
    if not os.path.exists(fldr):
        os.makedirs(fldr)

    fig = plt.figure(figsize=(16,12));
    cmap=plt.get_cmap('tab20')

    for model_n in range(2):
        model_dir = os.path.join(p,'data/rnn/multitask/',dir_specific_all,'most',str(model_n))

        h_tf = {}
        hiddens = {}
    #     T = 100
    #     print(T)
    
        trial_temp = gen_trials_from_model_dir(model_dir,'delaygo')
        
        for rule, task_num in zip(rule_set,task_num_set):
            trial = trial_temp
            trial.x[:,:,5:] = 0
            trial.x[:,:,5+task_num] = 1
            ckpt_n = name_best_ckpt(model_dir,'delaygo')
            ckpt_n_dir = os.path.join(model_dir,'ckpts/model.ckpt-' + str(int(ckpt_n)))
            X, x = gen_X_from_model_dir_epoch(model_dir,ckpt_n_dir,trial,'stim1')
            h_tf[rule] = np.transpose(x,(2,1,0))                   
            T,_,N = np.shape(h_tf[rule])
            t_use = T
            hiddens[rule] = np.reshape(h_tf[rule][t_use-1:t_use,0:200:2,:],(-1, N))  # was batch x time x dim

        evr = {}
        r1 = 0
        for rule1 in rule_set:
            r1 +=1
            ax = fig.add_subplot(2, 6, r1)
            pca = PCA(n_components=n_components).fit(hiddens[rule1])

            for rule2, task_num in zip(rule_set,task_num_set):
                h_pca = pca.transform(hiddens[rule2])
                evr[rule1,rule2] = np.var(h_pca,axis=0)/np.sum(np.var(h_pca,axis=0))
                plt.plot(np.cumsum(evr[rule1,rule2]), '-o', c = cmap(task_num/max(task_num_set)))

            plt.title(rule1 + '_' + dir_specific_all)
            plt.xlim((-1,n_components))

            if r1==1:
                plt.legend(rule_set)
                plt.xlabel('# Components')
                plt.ylabel('Cumulative Variance Explained')
                

    save_dir = fldr + '/pc_across_' + dir_specific_all
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)
    plt.savefig(save_dir + '_time' + str(t_use)+'_model_' + str(model_n) + '.svg')
    plt.show()
```

```python
def vanilla_run_with_h0(params, x_t, h0):
    h = h0
    h_t = []
    h_t.append(h)
    for x in x_t:
        h = rnn_vanilla(params, np.squeeze(h), np.squeeze(x.T))
        h_t.append(np.expand_dims(h,axis=1))

    h_t = np.squeeze(np.array(h_t))  
    o_t = batch_affine(params, h_t)
    return h_t, o_t
```
