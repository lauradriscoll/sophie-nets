```python
def dim_prep(X):
    # Split each dataset into two halves: training set and test set
    nSamples = int(np.shape(X)[0]/2)
    train = X[0:nSamples,:]
    test = X[nSamples:2*nSamples,:]
    return train, test

def dim_compute(X):
    train = X# X[0:int(X.shape[0]/2),:]
    test = X# X[int((X.shape[0]/2)):X.shape[0],:]

    pca = PCA()
    pca.fit(train)
    ev_train = pca.explained_variance_
    ev_test = np.var(pca.transform(test),axis=0)
    return ev_test

def gen_trials_from_model_dir(model_dir,rule):
    model = Model(model_dir)
    with tf.Session() as sess:
        model.restore()
# # # #         model.cell._sigma=0
        # get all connection weights and biases as tensorflow variables
        var_list = model.var_list
        # evaluate the parameters after training
#         params = [sess.run(var) for var in var_list]
        # get hparams
        hparams = model.hp
        # create a trial
        trial = generate_trials(rule, hparams, mode='random', noise_on=False, batch_size =500, delay_fac =1)
    return trial  

def gen_single_trial_from_model_dir(model_dir,rule):
    model = Model(model_dir)
    with tf.Session() as sess:
        model.restore()
#         model.cell._sigma=0
        # get all connection weights and biases as tensorflow variables
        var_list = model.var_list
        # evaluate the parameters after training
#         params = [sess.run(var) for var in var_list]
        # get hparams
        hparams = model.hp
        # create a trial
        trial = generate_trials(rule, hparams, mode='random', noise_on=False, batch_size =1, delay_fac =1)
    return trial    

def load_X_from_model_dir_oc(model_dir,trial):
    model = Model(model_dir)
    with tf.Session() as sess:
        model.restore()
# # #         model.cell._sigma=0
        # get all connection weights and biases as tensorflow variables
        var_list = model.var_list
        # evaluate the parameters after training
        params = [sess.run(var) for var in var_list]
        w_out = params[2]
        oc = orthogonal_complement(w_out)
        hparams = model.hp
        feed_dict = tools.gen_feed_dict(model, trial, hparams)
        # run model
        h_tf, y_hat_tf = sess.run([model.h, model.y_hat], feed_dict=feed_dict) #(n_time, n_condition, n_neuron)
        x = np.transpose(h_tf,(2,1,0))
        X = np.reshape(x,(x.shape[0],-1))
        oc_X = np.matmul(X.T, oc)
    return oc_X    #return orthogonal complement of hidden unit activity to ouput projection matrix

def load_X_from_model_dir(model_dir,trial):
    model = Model(model_dir)
    with tf.Session() as sess:
        model.restore()
#         model.cell._sigma=0
        # get all connection weights and biases as tensorflow variables
        var_list = model.var_list
        # evaluate the parameters after training
        hparams = model.hp
        feed_dict = tools.gen_feed_dict(model, trial, hparams)
        # run model
        h_tf, y_hat_tf = sess.run([model.h, model.y_hat], feed_dict=feed_dict) #(n_time, n_condition, n_neuron)
        x = np.transpose(h_tf,(2,1,0))
        X = np.reshape(x,(x.shape[0],-1))
    return X    #return orthogonal complement of hidden unit activity to ouput projection matrix

def gen_X_from_model_dir_epoch(model_dir,rule,epoch,range_size):
    model = Model(model_dir)
    with tf.Session() as sess:
        model.restore()
# #         model.cell._sigma=0
        # get all connection weights and biases as tensorflow variables
        var_list = model.var_list
        # evaluate the parameters after training
        params = [sess.run(var) for var in var_list]
        # get hparams
        hparams = model.hp
        # create a trial
        trial = generate_trials(rule, hparams, mode='test', noise_on=False, batch_size =500)        
        feed_dict = tools.gen_feed_dict(model, trial, hparams)
        # run model
        h_tf, y_hat_tf = sess.run([model.h, model.y_hat], feed_dict=feed_dict) #(n_time, n_condition, n_neuron)
        if trial.epochs[epoch][1] is None:
            epoch_range = range(np.shape(h_tf)[0]-range_size,np.shape(h_tf)[0])
        else:
            epoch_range = range(trial.epochs[epoch][1]-range_size,trial.epochs[epoch][1])
        x = np.transpose(h_tf[epoch_range,:,:],(2,1,0))
        X = np.reshape(x,(x.shape[0],-1))
    return X, trial    #return hidden unit activity

def load_X3_from_model_dir(model_dir,trial):
    model = Model(model_dir)
    with tf.Session() as sess:
        model.restore()
#         model.cell._sigma=0
        # get all connection weights and biases as tensorflow variables
        var_list = model.var_list
        # evaluate the parameters after training
        params = [sess.run(var) for var in var_list]
        w_out = params[2]
        hparams = model.hp
        feed_dict = tools.gen_feed_dict(model, trial, hparams)
        # run model
        h_tf, y_hat_tf = sess.run([model.h, model.y_hat], feed_dict=feed_dict) #(n_time, n_condition, n_neuron)
        x3 = np.transpose(h_tf,(2,1,0))
    return x3    #return orthogonal complement of hidden unit activity to ouput projection matrix

def find_X_corr_all(model_dir,rule,trial,epoch_name,epoch_bounds,n_models=20):

    trial_size = int(trial.y.shape[0])
    epoch_size = int(epoch_bounds[1]-epoch_bounds[0]) 

    X_corr = np.zeros((n_models,trial_size,epoch_size,epoch_size))
    epoch_inds = range(epoch_bounds[0],epoch_bounds[1])

    for model_n in range(0,n_models):

        model_path = os.path.join(model_dir,str(model_n))
        x3 = load_X3_from_model_dir(model_path,trial)

        t_ind = 0
        for t in range(0,trial_size):
            X_corr[model_n,t_ind,:,:] = np.corrcoef(x3[:,t,epoch_inds].T)
            t_ind+=1

    return X_corr

def find_X_corr_single(model_dir,rule,trial,epoch_name,epoch_bounds,n_models=20):

    trial_size = int(trial.y.shape[0])
    epoch_size = int(epoch_bounds[1]-epoch_bounds[0]) 

    X_corr = np.zeros((n_models,trial_size,epoch_size,epoch_size))
    epoch_inds = range(epoch_bounds[0],epoch_bounds[1])

    for model_n in range(0,n_models):

        model_path = os.path.join(model_dir,rule,str(20+model_n))
        x3 = load_X3_from_model_dir(model_path,trial)

        t_ind = 0
        for t in range(0,trial_size):
            X_corr[model_n,t_ind,:,:] = np.corrcoef(x3[:,t,epoch_inds].T)
            t_ind+=1

    return X_corr
```

```python
task1 = 'reactgo'
task2 = 'reactanti'

ve_rand = {}
ve_rrand = {}
ve_out = {}
ve_in = {}
n_components = 20#X_all.shape[0]

for model_n in range(0,20):
    tf.reset_default_graph()
    model_path_all = os.path.join('/home/laura/data/multitask/train_all',str(model_n))
    model = Model(model_path_all)
    with tf.Session() as sess:
        model.restore()
#         model.cell._sigma=0
        # get all connection weights and biases as tensorflow variables
        var_list = model.var_list
        # evaluate the parameters after training
        params = [sess.run(var) for var in var_list]
        w_out = params[2]
        hparams = model.hp    
        trial = generate_trials(task1, hparams, mode='random', noise_on=False, batch_size =500)
        feed_dict = tools.gen_feed_dict(model, trial, hparams)
        # run model
        h_tf, y_hat_tf = sess.run([model.h, model.y_hat], feed_dict=feed_dict) #(n_time, n_condition, n_neuron)
        x = np.transpose(h_tf,(2,1,0))
        X_go = np.reshape(x,(x.shape[0],-1))

        trial = generate_trials(task2, hparams, mode='random', noise_on=False, batch_size =500)
        feed_dict = tools.gen_feed_dict(model, trial, hparams)
        # run model
        h_tf, y_hat_tf = sess.run([model.h, model.y_hat], feed_dict=feed_dict) #(n_time, n_condition, n_neuron)
        x = np.transpose(h_tf,(2,1,0))
        X_anti = np.reshape(x,(x.shape[0],-1))
        
        rule_train_now = hparams['rng'].choice(hparams['rule_trains'],p=hparams['rule_probs'])
        trial = generate_trials(rule_train_now,hparams, mode='random', noise_on=False, batch_size =500)
        feed_dict = tools.gen_feed_dict(model, trial, hparams)
        # run model
        h_tf, y_hat_tf = sess.run([model.h, model.y_hat], feed_dict=feed_dict) #(n_time, n_condition, n_neuron)
        x = np.transpose(h_tf,(2,1,0))
        X_all = np.reshape(x,(x.shape[0],-1))        
        
    go_row_means = np.mean(X_go, axis=1)
    X_go_center = X_go - go_row_means.reshape(X_go.shape[0],1)
    
    anti_row_means = np.mean(X_anti, axis=1)
    X_anti_center = X_anti - anti_row_means.reshape(X_anti.shape[0],1)
    
#     X_all = np.concatenate((X_go, X_anti), axis=1)
    all_row_means = np.mean(X_all, axis=1)
    X_all_center = X_all - all_row_means.reshape(X_all.shape[0],1)
        
#     X_go = StandardScaler().fit_transform(X_go.T, 'with_mean' = True, 'with_std' = False).T
#     X_anti = StandardScaler().fit_transform(X_anti.T, 'with_mean' = True, 'with_std' = False).T
#     X_all = StandardScaler().fit_transform(X_all.T, 'with_mean' = True, 'with_std' = False).T
        
    C = np.cov(X_all_center)
    S, U = LA.eig(C)
    rand_mat = npr.randn(X_all_center.shape[0],n_components)

    numer = np.dot(np.dot(U,np.diag(S**(1/2))),rand_mat)
    denom = LA.norm(np.dot(np.dot(U,np.diag(S**(1/2))),rand_mat))
    v_align = LA.orth(numer/denom).T
    #project anti data into random subspace
    pca_rand = np.dot(v_align,X_go_center)
    pca_rrand = np.dot(LA.orth(rand_mat).T,X_go_center)
    
    pca = PCA(n_components=n_components) 
    pca.fit(X_go_center.T)
    pca_in = pca.transform(X_go_center.T).T
    
    pca = PCA(n_components=n_components)
    pca.fit(X_anti_center.T)
    pca_out = pca.transform(X_go_center.T).T
    
    ve_in[model_n] = np.cumsum(np.var(pca_in,axis=1)/sum(np.var(X_go_center,axis=1)))
    ve_out[model_n] = np.cumsum(np.var(pca_out,axis=1)/sum(np.var(X_go_center,axis=1)))
    ve_rand[model_n] = np.cumsum(np.var(pca_rand,axis=1)/sum(np.var(X_go_center,axis=1)))
    ve_rrand[model_n] = np.cumsum(np.var(pca_rrand,axis=1)/sum(np.var(X_go_center,axis=1)))
```
