```python
all_rules = ['fdgo', 'reactgo', 'delaygo', 'fdanti', 'reactanti', 'delayanti',
              'dm1', 'dm2', 'contextdm1', 'contextdm2', 'multidm',
              'delaydm1', 'delaydm2', 'contextdelaydm1', 'contextdelaydm2', 'multidelaydm',
              'dmsgo', 'dmsnogo', 'dmcgo', 'dmcnogo']

def gen_trials_from_model_dir(model_dir,rule):
    model = Model(model_dir)
    with tf.Session() as sess:
        model.restore()
        model._sigma=0
        # get all connection weights and biases as tensorflow variables
        var_list = model.var_list
        # evaluate the parameters after training
#         params = [sess.run(var) for var in var_list]
        # get hparams
        hparams = model.hp
        # create a trial
        trial = generate_trials(rule, hparams, mode='test', noise_on=False, batch_size =1000, delay_fac =1)
    return trial  

def load_X_from_model_dir(model_dir,d,trial,epoch):
    model = Model(model_dir)
    with tf.Session() as sess:
        model.saver.restore(sess,d)
        model._sigma=0
        # get all connection weights and biases as tensorflow variables
        var_list = model.var_list
        # evaluate the parameters after training
        hparams = model.hp
        feed_dict = tools.gen_feed_dict(model, trial, hparams)
        # run model
        h_tf, y_hat_tf = sess.run([model.h, model.y_hat], feed_dict=feed_dict) #(n_time, n_condition, n_neuron)
        x = np.transpose(h_tf[trial.epochs[epoch][1]-1,:,:],(1,0)) # h_tf[:,range(1,n_trials),:],(2,1,0))
        X = np.reshape(x,(x.shape[0],-1))
    return X    #return orthogonal complement of hidden unit activity to ouput projection matrix

def restore_ckpt(model_dir, ckpt_n):
    ckpt_n_dir = os.path.join(model_dir,'ckpts/model.ckpt-' + str(int(ckpt_n)) + '.meta')
    model = Model(model_dir)
    with tf.Session() as sess:
        model.saver.restore(sess,ckpt_n_dir)
    return model

def find_ckpts(model_dir):
    s_all = []
    ckpt_n_dir = os.path.join(model_dir,'ckpts/')
    for file in os.listdir(ckpt_n_dir):
        if file.endswith('.meta'):
            m = re.search('model.ckpt(.+?).meta', file)
            if m:
                found = m.group(1)
            s_all = np.concatenate((s_all,np.expand_dims(abs(int(found)),axis=0)),axis = 0)
    return s_all.astype(int)

def name_best_ckpt(model_dir,rule):
    s_all = find_ckpts(model_dir)
    s_all_inds = np.sort(s_all/1000)
    s_all_inds = s_all_inds.astype(int)
    fname = os.path.join(model_dir, 'log.json')
    
    with open(fname, 'r') as f:
        log_all = json.load(f)
        x = log_all['cost_'+rule]
           
    y = [x[j] for j in s_all_inds]
    ind = int(1000*s_all_inds[np.argmin(y)])
    return ind

def plot_stim_rep_fit(X):
    
    N,S = np.shape(X) # neurons x trials (stim) 
    
    pca = PCA()
    X_pca = pca.fit_transform(X.T)
    ev_all = pca.explained_variance_ratio_
    
    plt.figure(figsize=(20,5))
    plt.rcParams.update({'font.size': 16})
    cmap=plt.get_cmap('rainbow')
    col = 2

    plt.subplot(1,col,1)
    plt.plot(np.cumsum(ev_all[0:10]/sum(ev_all)),'-ok')

    for s in range(S):
        c = cmap(s/S)

        plt.subplot(1,col,2)
        plt.plot(X_pca.T[0,s],X_pca.T[1,s],'.',c = c)
        plt.axis('equal')
        plt.xlabel('PC1')
        plt.ylabel('PC2')
        plt.plot()
        
    return pca
        
def plot_stim_rep_transform(X,pca):
    
    N,S = np.shape(X) # neurons x trials (stim) 
    
    X_pca = pca.transform(X.T)
    ev_all = pca.explained_variance_ratio_
    
    plt.figure(figsize=(20,5))
    plt.rcParams.update({'font.size': 16})
    cmap=plt.get_cmap('rainbow')
    col = 2

    plt.subplot(1,col,1)
    plt.plot(np.cumsum(ev_all[0:10]/sum(ev_all)),'-ok')

    for s in range(S):
        c = cmap(s/S)

        plt.subplot(1,col,2)
        plt.plot(X_pca.T[0,s],X_pca.T[1,s],'.',c = c)
        plt.axis('equal')
        plt.xlabel('PC1')
        plt.ylabel('PC2')
        plt.plot()
        
def visualize_data_procrustes(data1, data2):
    N,S = np.shape(data1)
    col = 2
    pca = plot_stim_rep_fit(np.concatenate((data1,data2),axis=1))
    # plot_stim_rep_transform(mtx2,pca)
    plt.subplot(1,col,1)
    plt.xlabel('# PCs')
    plt.ylabel('Variance Explained')

    X1_pca = pca.transform(data1.T)
    X2_pca = pca.transform(data2.T)

    for s in range(S):
        c = cmap(s/S)

        plt.subplot(1,col,2)
        plt.plot(X1_pca.T[0,s],X1_pca.T[1,s],'.',c = c)
        plt.plot(X2_pca.T[0,s],X2_pca.T[1,s],'+',c = c)
        plt.axis('equal')
        plt.xlabel('PC1')
        plt.ylabel('PC2')

    plt.show()
    return pca
    
def vsr(task_num, model_n, single, dir_specific):
    r"""Visualize Stimulus Representation
  
    """
    epoch = 'stim1'
    cmap=plt.get_cmap('rainbow')
    rule = rules_dict['all'][task_num]
    
    if single:
        model_dir_all = os.path.join('/Users/lauradriscoll/Documents/data/rnn/multitask/lowD/',dir_specific,rule,str(model_n))
        
        file_name = 'Single Task Network lowD ' + rule + '_' + str(model_n) 
    else:
        model_dir_all = os.path.join('/Users/lauradriscoll/Documents/data/rnn/multitask/lowD/',dir_specific,'most',str(model_n))
        file_name = 'Multitask Network lowD ' + rule + '_' + str(model_n) 
        
    trial_temp = gen_trials_from_model_dir(model_dir_all,'delaygo')
    trial = trial_temp
    trial.x[:,:,5:] = 0
    trial.x[:,:,5+task_num] = 1

    model = Model(model_dir_all)
    ckpt_n = name_best_ckpt(model_dir_all,rule)
    ckpt_n_dir = os.path.join(model_dir_all,'ckpts/model.ckpt-' + str(int(ckpt_n)))
    X1 = load_X_from_model_dir(model_dir_all,ckpt_n_dir,trial,epoch)
    X = X1[:,::2]
    N,S = np.shape(X) # neurons x trials (stim)  
    
    #visualize stim on all trials
    FONT_WEIGHT = 'bold'
    fig = plt.figure(figsize=(5, 5),tight_layout=True,facecolor='white')
    plt.rcParams.update({'font.size': 20})
    ax = plt.subplot(1,1,1)
    plt.plot(X.T)
    plt.xticks(range(0,S,int(S/4)), [int(i) for i in range(0,361,90)])


#     ax.set_ylabel('Activation', fontweight=FONT_WEIGHT)
#     ax.set_xlabel('Stimulus Angle', fontweight=FONT_WEIGHT)
#     ax.legend(('Hidden Unit 1','Hidden Unit 2', '...'))
    plt.text(0,0,file_name + '_' + dir_specific)
    plt.savefig('cosyne/transform_stim_rep/' + file_name + '_' + dir_specific + '_activations.svg')
    
    #visualize stim rep in PC space
    col = 4 
    pca = PCA()
    X_pca = pca.fit_transform(X.T)
    ev_all = pca.explained_variance_ratio_

    fig = plt.figure(figsize=(15, int(15/col)),tight_layout=True,facecolor='white')
    plt.rcParams.update({'font.size': 16})
    cmap=plt.get_cmap('rainbow')

    plt.subplot(1,col,1)
    
    plt.plot(np.cumsum(ev_all[0:10]/sum(ev_all)),'-ok')
#     plt.xlabel('Number of PCs')
#     plt.ylabel('Variance Explained')
    plt.ylim([0, 1.1])
    
    plt.text(0,0,file_name + '_' + dir_specific)
    plt.xticks(range(0,10,2), [int(i) for i in range(1,11,2)])

    for s in range(0,S,1):
        c = cmap(s/S)

        for x in range(1,col):
            plt.subplot(1,col,x+1)
            plt.plot(X_pca.T[2*x-2,s],X_pca.T[2*x-1,s],'.',c = c)
#             plt.plot(X_pca.T[2*x-2,s+1],X_pca.T[2*x-1,s+1],'o',c = c, markerfacecolor = 'none')
            plt.axis('equal')
            plt.xlabel('PC'+str(2*x-1))
            plt.ylabel('PC'+str(2*x))
            plt.plot()

    plt.savefig('cosyne/transform_stim_rep/' + file_name + '_' + dir_specific + '_stim_rep.svg')
```

```python
def prep_procrustes(data1, data2):
    r"""Procrustes analysis, a similarity test for two data sets.
    
    Parameters
    ----------
    data1 : array_like
        Matrix, n rows represent points in k (columns) space `data1` is the
        reference data, after it is standardised, the data from `data2` will be
        transformed to fit the pattern in `data1` (must have >1 unique points).
    data2 : array_like
        n rows of data in k space to be fit to `data1`.  Must be the  same
        shape ``(numrows, numcols)`` as data1 (must have >1 unique points).
    Returns
    -------
    mtx1 : array_like
        A standardized version of `data1`.
    mtx2 : array_like
        The orientation of `data2` that best fits `data1`. Centered, but not
        necessarily :math:`tr(AA^{T}) = 1`.
    disparity : float
        :math:`M^{2}` as defined above.
  
    """
    mtx1 = np.array(data1, dtype=np.double, copy=True)
    mtx2 = np.array(data2, dtype=np.double, copy=True)

    if mtx1.ndim != 2 or mtx2.ndim != 2:
        raise ValueError("Input matrices must be two-dimensional")
    if mtx1.shape != mtx2.shape:
        raise ValueError("Input matrices must be of same shape")
    if mtx1.size == 0:
        raise ValueError("Input matrices must be >0 rows and >0 cols")

    # translate all the data to the origin
    mtx1 -= np.mean(mtx1, 0)
    mtx2 -= np.mean(mtx2, 0)

    norm1 = np.linalg.norm(mtx1)
    norm2 = np.linalg.norm(mtx2)

    if norm1 == 0 or norm2 == 0:
        raise ValueError("Input matrices must contain >1 unique points")

    # change scaling of data (in rows) such that trace(mtx*mtx') = 1
    mtx1 /= norm1
    mtx2 /= norm2
    
    return mtx1,mtx2

def procrustes(mtx1, mtx2):
    # transform mtx2 to minimize disparity
    R, s = orthogonal_procrustes(mtx1, mtx2)
    mtx2 = np.dot(mtx2, R.T) * s

    # measure the dissimilarity between the two datasets
    disparity = np.sum(np.square(mtx1 - mtx2))

    return mtx1, mtx2, disparity, R, s
```
