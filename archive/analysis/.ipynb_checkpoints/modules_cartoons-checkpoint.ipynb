```python
from __future__ import absolute_import, division, print_function
import glob
import os
import pdb
import numpy as np
import numpy.random as npr
import tensorflow as tf
import sys
import matplotlib.pyplot as plt
import pdb
import getpass
from scipy import stats
from sklearn import linear_model
from scipy import linalg as LA
import numpy.random as npr
from sklearn.decomposition import PCA
```

```python
N = 256
N_a = 150
N_b = N-N_a

W = np.zeros((N,N)) #npr.randn(N,N)
W[:N_a,:N_a] = npr.randn(N_a,N_a)+1.5
W[N_a:,N_a:] = npr.randn(N_b,N_b)+1.5
```

```python
plt.hist(W.flatten())
plt.show()
```

```python
fig = plt.figure(figsize=(7,5),tight_layout=True,facecolor='white')
plt.rcParams.update({'font.size': 20})
ax = plt.subplot(1,1,1)
plt.imshow(W,cmap = 'PuOr')
plt.xlabel('Pre '+r'$cell_j$')
plt.ylabel('Post '+r'$cell_i$')
plt.clim(-2,2)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.colorbar()

plt.show()
```

```python
A = npr.randn(N,N)
R = LA.orth(A)
RWRT = np.matmul(R,np.matmul(W,R.T))

fig = plt.figure(figsize=(7,5),tight_layout=True,facecolor='white')
plt.rcParams.update({'font.size': 20})
ax = plt.subplot(1,1,1)
plt.imshow(RWRT,cmap = 'PuOr')
plt.xlabel('Pre '+r'$cell_j$')
plt.ylabel('Post '+r'$cell_i$')
plt.clim(-2,2)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.colorbar()
plt.show()
```
