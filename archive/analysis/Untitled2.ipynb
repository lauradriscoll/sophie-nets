```python
import numpy as np
import matplotlib.pyplot as plt
import getpass
import os

ui = getpass.getuser()
if ui == 'laura':
    p = '/home/laura'
elif ui == 'lauradriscoll':
    p = '/Users/lauradriscoll/Documents'
elif ui == 'lndrisco':
    p = '/home/users/lndrisco'
    
m = os.path.join(p,'code/multitask-nets/','schematics')
figpath = os.path.join(m,'Dynamics Schematics')
if not os.path.exists(figpath):
    os.makedirs(figpath)
```

```python
#params

xmin = -5
xmax = 5
num_arrows = 10
scale = 11

x,y = np.meshgrid(np.linspace(xmin,xmax,num_arrows),np.linspace(xmin,xmax,num_arrows))
```

```python
u = x/np.sqrt(x**2 + y**2)
v = y/np.sqrt(x**2 + y**2)

plt.quiver(x,y,u,v,scale=scale)
plt.axis('off')
plt.axis('square')

figname = 'expand.pdf'
plt.savefig(os.path.join(figpath,figname))
plt.show()
```

```python
#params

xmin = -5
xmax = 5
num_arrows = 6
scale = 5

x,y = np.meshgrid(np.linspace(xmin,xmax,num_arrows),np.linspace(xmin,xmax,num_arrows))

x1 = x+.1
y1 = y+0

for alpha in np.linspace(0,1,5):
    
    u_m1 = (1-alpha)*x1/np.sqrt(x1**2 + y1**2)
    v_m1 = (1-alpha)*y1/np.sqrt(x1**2 + y1**2)
    u_m2 = alpha*x/np.sqrt(x**2 + y**2)
    v_m2 = alpha*y/np.sqrt(x**2 + y**2)

    plt.quiver(x1,y1,u_m1,v_m1, color = 'm',scale=scale)
    plt.quiver(x,y,u_m2,v_m2, color = 'c',scale=scale)
    plt.axis('off')
    plt.axis('square')

    figname = str(alpha*10)+'expand.pdf'
    plt.savefig(os.path.join(figpath,figname))
    plt.show()
```
